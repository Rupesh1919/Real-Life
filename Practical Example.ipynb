```python
cols=['Mileage', 'EngineV', 'log_price', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed=data_with_dummies[cols]
```

```python
data_preprocessed.head()
```

Linear Regression

DECLARE THE INPUTS AND TARGETS

```python
target=data_preprocessed['log_price']
```

```python
inputs=data_preprocessed.drop('log_price', axis=1)
```

SCALE THE DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
input_scaled_data=scaler.transform(inputs)
```

TRAIN TEST AND SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_scaled_data,target,test_size=0.2,random_state=365)
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_hat=reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets(y_train)', size=10)
plt.ylabel('Predictions(yhat)',size=10)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

pdf

```python
sns.distplot(y_train-y_hat)
plt.title('Residual Pdf', size=10)
```

```python
reg.score(x_train,y_train)
```

```python
reg.intercept_
```

```python
reg_summary=pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights']=reg.coef_
```

```python
reg_summary
```

    Testing

```python
y_hat_test=reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('Targets(y_test)', size=10)
plt.ylabel('Predictions(y_hat_test)',size=10)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf=pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
```

```python
df_pf
```

```python
df_pf['Target']=np.exp(y_test)
```

```python
df_pf
```

```python
y_test=y_test.reset_index(drop=True)
```
