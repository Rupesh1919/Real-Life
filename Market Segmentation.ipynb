Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

Load the data

```python
data=pd.read_csv(r"D:/The Data Science Course 2020 - All Resources/Part_5_Advanced_Statistical_Methods_(Machine_Learning)/S38_L265/3.12. Example.csv")
data
```

Select the features

```python
x=data.copy()
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
clusters=x.copy()
clusters['Clusters']=kmeans.fit_predict(x)
clusters
```

Plot the data

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['Clusters'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyality')
```

scaling the data

```python
from sklearn import preprocessing
```

```python
x_scaled=preprocessing.scale(x)
x_scaled
```

Advantages of clusters

```python
wcss=[]
for i in range(1,10):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

Explore the clusters and select the number

```python
kmeans=KMeans(4)
kmeans.fit(x_scaled)
```

```python
cluster_new=x.copy()
cluster_new['Clusters']=kmeans.fit_predict(x_scaled)
cluster_new
```

Plot the data

```python
plt.scatter(cluster_new['Satisfaction'],cluster_new['Loyalty'],c=cluster_new['Clusters'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```
